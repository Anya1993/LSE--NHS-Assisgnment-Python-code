**Question 2:** Which service setting was the most popular for NHS North West London from 1 January to 1 June 2022?

```python
# For each of these service settings, determine the number of records available for the period and the location.Â 
# Count records per service setting and location

# Filter by location
nc_subset = nc[nc['sub_icb_location_name'] == 'NHS North West London ICB - W2U3Z']

# Define date range
start_date = '2022-01-01'
end_date = '2022-06-01'

# Apply date filter
dates_filter = nc_subset[
    (nc_subset['appointment_date'] >= start_date) & 
    (nc_subset['appointment_date'] <= end_date)]

# Group by service_setting and sum appointments
ss_cnt = dates_filter.groupby('service_setting', as_index=False)['count_of_appointments'].sum()

# Rename column for clarity
ss_cnt = ss_cnt.rename(columns={'count_of_appointments': 'total_appointments'})

# View result
print(ss_cnt.head())
```

```python
# Create a plot to forecast the appointments based on ar data set with ARIMA
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

ar_2 = ar.copy()

ar_2['appointment_month'] = pd.to_datetime(ar_2['appointment_month'])
ar_2.set_index('appointment_month', inplace=True)

ar_monthly = ar_2['count_of_appointments'].resample('M').sum()
ar_monthly

model = ARIMA(ar_monthly, order=(1, 1, 1))
model_fit = model.fit()

# Print summary of the model
print(model_fit.summary())

# Plot the forecasts
forecast = model_fit.forecast(steps=6)
plt.figure(figsize=(10, 6))
plt.plot(ar_monthly, label='Original Monthly Appointments')
plt.plot(forecast, label='Forecast', color='orange')
plt.title("The appointments will tend to increase from the forecast", fontsize=10)
plt.suptitle("Monthly Appointments with ARIMA Forecast", fontsize=16)
plt.xlabel('Month and Year')
plt.ylabel('Number of appointments')
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plt.legend()
plt.show()

forecast_values = model_fit.forecast(steps=6)
forecast_values
```
